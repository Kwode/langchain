```python
from langchain_core.prompts import ChatPromptTemplate, HumanMessagePromptTemplate, AIMessagePromptTemplate, FewShotChatMessagePromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.messages import HumanMessage
```

```python
template_s = '{system description}'
template_h = 'could you give me a nickname like {name}?'

message_template_s = SystemMessagePromptTemplate.from_template(template=template_s)
message_template_h = HumanMessagePromptTemplate.from_template(template=template_h)
```

```python
chat_template = ChatPromptTemplate.from_messages([message_template_s, message_template_h])
chat_template

chat_value = chat_template.invoke({'system description':'You are a name smith', 'name':'Badoink'})
```

```python
chat = ChatOpenAI(
    model='gpt-3.5-turbo',
    temperature=1,
    max_completion_tokens=20
)

res=chat.invoke(chat_value)

print(res.content)
```

FEW-SHOT PROMPTING

```python
human_message = 'could you give me a nickname from {name}?'
ai_response = '{response}'

human_message_template = HumanMessagePromptTemplate.from_template(template=human_message)
ai_message_template = AIMessagePromptTemplate.from_template(template=ai_response)

main_chat_template = ChatPromptTemplate.from_messages([human_message_template, ai_message_template])

prompts = [
    {'name':'Fred', 'response':'Sure! How about "Fredorino"?'},
    {'name':'Machavelli', 'response':'Sure! How about "Machavells"?'},
]

few_shots = FewShotChatMessagePromptTemplate(examples=prompts, example_prompt=main_chat_template, input_variables=['name'])

few_shot_chat_template = ChatPromptTemplate.from_messages([few_shots, human_message])
few_shot_value = few_shot_chat_template.invoke({'name':'Dora'})
few_shot_value
```

```python
response = chat.invoke(few_shot_value)
for i in few_shot_value.messages:
    print(f"{i.type}: {i.content}\n")

print(f"ai: {response.content}")
```

STRING OUTPUT PARSER

```python
human_message_2  = HumanMessage(content="tell me something interesting about a dog")

response = chat.invoke([human_message_2])

str_parser = StrOutputParser()
print(str_parser.invoke(response))
```

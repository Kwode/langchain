```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import CommaSeparatedListOutputParser
```

```python
model = ChatOpenAI(
    model='gpt-3.5-turbo',
    temperature=0.7,
    max_completion_tokens=20
)
```

```python
list_parser = CommaSeparatedListOutputParser()
```

```python
chat_template = ChatPromptTemplate.from_messages([('human', 'give me a list of 3 {pet} breeds. '+ CommaSeparatedListOutputParser().get_format_instructions())])
```

```python
chain = chat_template | model | list_parser
chain.invoke({'pet':'dog'})
```

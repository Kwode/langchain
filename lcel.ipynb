```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import CommaSeparatedListOutputParser
```

```python
model = ChatOpenAI(
    model='gpt-3.5-turbo',
    temperature=0.7,
    max_completion_tokens=20
)
```

```python
list_parser = CommaSeparatedListOutputParser()
```

Batching

```python
batch_chat_template = ChatPromptTemplate.from_messages([('human', 'I just got a {object} for {sport}, Could you suggest training tips?')])
```

```python
chain = batch_chat_template | model
```

```python
chain.batch([{'object':'ball', 'sport':'basketball'}, {'object':'racket', 'sport':'tennis'}])
```

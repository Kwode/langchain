```python
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
```

```python
load_dotenv()

TEMPLATE = """
System:
{description}

Human:
I've recently adopted a {pet}. could you suggest a {pet} name?
"""
```

```python
prompt_template = PromptTemplate.from_template(template=TEMPLATE)
prompt_template
```

```python
prompt_value = prompt_template.invoke({'description': 'You are a well known name smith', 'pet':'cat'})
```

```python
prompt_value
```

```python
llm = ChatOpenAI(
    model='gpt-3.5-turbo',
    temperature=0.7,
    max_completion_tokens=50
)

response = llm.invoke(prompt_value)
print(response.content)
```

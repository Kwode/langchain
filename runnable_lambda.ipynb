```python
from langchain_core.runnables import RunnableLambda, chain
```

```python
runnable_sum = RunnableLambda(lambda x: sum(x))
runnable_sq = RunnableLambda(lambda x: x**2)
math_chain = runnable_sum | runnable_sq
num = [1, 2, 3, 4, 5]
hum = [6, 7, 8, 9, 10]
new_list = num + hum
math_chain.invoke(new_list)
```

```python
def find_sum(x):
    return sum(x)

def find_sq(x):
    return x**2

math_chain_2 = RunnableLambda(find_sum) | RunnableLambda(find_sq)
math_chain_2.invoke(new_list)
```

@chain decorator

```python
@chain
def find_sum(x):
    return sum(x)

@chain
def find_sq(x):
    return x**2

math_chain_3 = find_sum | find_sq

math_chain_3.invoke(new_list)
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
```

```python
model = ChatOpenAI(
    model='gpt-3.5-turbo', 
    temperature=0.7,
    max_completion_tokens=50
)
```

```python
chat_1_template = ChatPromptTemplate.from_template("""
list 5 {color} fruits in {country}
""")
```

```python
str_parser = StrOutputParser()
```

```python
chain_1 = chat_1_template | model | str_parser | {'fruits':RunnablePassthrough(), 'color':RunnablePassthrough()}
```

```python
chat_2_template = ChatPromptTemplate.from_template("""
why is {fruits} {color}?
""")
```

```python
chain_2 =  chat_2_template | model | str_parser
```

```python
chain_combined = chain_1 | chain_2
chain_combined.invoke({'country':'algeria', 'color':'red'})
```

```python
chain_combined.get_graph().print_ascii()
```
